```python
#Pie Chart
import matplotlib.pyplot as plt

#Data Source https://gs.statcounter.com/vendor-market-share/mobile/pakistan
corporations = ['Samsung', 'Vivo', 'Infinix', 'Oppo', 'Tecno', 'Xiaomi', 'Others']
market_share = [17.92, 13.89, 13.67, 13.34, 8.6, 6.6, 25.98]  # percentages

# Colors for each slice of the pie chart
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

# Exploding the 4th slice (optional, to highlight a specific corporation)
explode = (0.15, 0, 0, 0, 0, 0, 0)

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(market_share, labels=corporations, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode)

# Draw circle
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add a title
plt.title('Market Share by Mobile Vendor')

# Show the pie chart
plt.show()

```

```python
#Area Chart
import altair as alt
from vega_datasets import data
unemp = data.unemployment_across_industries()
alt.Chart(unemp).mark_area().encode(
    alt.X('date:T', axis=alt.Axis(format='%Y', labelAngle=0)),
    y='sum(count):Q'
)
hover = alt.selection_single(on='mouseover')

alt.Chart(unemp).mark_area().encode(
    alt.X('date:T', timeUnit='yearmonth',
      axis=alt.Axis(format='%Y', labelAngle=0, title='year')),
    alt.Y('sum(count)',
        stack='center'
    ),
    color=alt.Color('series:N', scale=alt.Scale(scheme="category20b")),
    opacity=alt.condition(hover, alt.value(1.0), alt.value(0.5))
).properties(
    selection=hover
)
```

```python
#Histogram
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.layouts import gridplot
import numpy as np

# Generate random data for the histogram
data = np.random.normal(0, 1, 1000)

# Create histogram
hist, edges = np.histogram(data, bins=20)

# Set up the Bokeh output file
output_file("histogram.html")

# Create a figure
p = figure(title="Histogram", background_fill_color="#E8DDCB")

# Define the edges of the bars
bottom = 0
top = hist
left = edges[:-1]
right = edges[1:]

# Add bars to the figure
p.quad(top=top, bottom=bottom, left=left, right=right, fill_color="navy", line_color="white", alpha=0.7)

# Show the plot
show(p)
```

```python
#Heat Map
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate a sample correlation matrix
np.random.seed(42)
data = np.random.rand(10, 10)
correlation_matrix = np.corrcoef(data, rowvar=False)

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)

# Add title
plt.title('Correlation Heatmap')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
x_data = np.arange(0, 10, 0.1)
y_data = np.sin(x_data)

line, = ax.plot(x_data, y_data)

def update(frame):
    line.set_ydata(np.sin(x_data + frame / 10.0))
    return line,

# Create the animation
animation = FuncAnimation(fig, update, frames=100, interval=100)

# Save the animation to a file
animation_file = "slow_animation.gif"
animation.save(animation_file, writer="imagemagick", fps=10)

# Display the saved animation
from IPython.display import Image
Image(filename=animation_file)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load a GeoDataFrame with a shapefile of world countries
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Create a variable for demonstration (e.g., population)
world['population'] = range(1, len(world) + 1)

# Normalize the population values for better visualization
world['normalized_population'] = world['population'] / world['population'].max()

# Create a cartogram using geopandas
fig, ax = plt.subplots(figsize=(15, 10))
world.boundary.plot(ax=ax, linewidth=1)
world.plot(column='normalized_population', ax=ax, legend=True, cmap='Blues', legend_kwds={'label': "Population by Country"})

# Add title
plt.title('Cartogram Example')

# Show the plot
plt.show()
```

```python
#Percentage Bar
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample data for percentage bar chart
categories = ['Category A', 'Category B', 'Category C', 'Category D']
counts = [30, 50, 20, 40]

# Calculate percentages
total = sum(counts)
percentages = [(count / total) * 100 for count in counts]

# Create a percentage bar chart using Seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x=categories, y=percentages, color='skyblue')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Percentage')
plt.title('Percentage Bar Chart Example')

# Show the plot
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 1), (1, 4)])

nx.draw(G, with_labels=True, font_weight='bold')
plt.title('Undirected Graph Visualization')
plt.show()
```

```python
#Directed Graph
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_edges_from([(1, 2), (2, 3), (3, 1), (1, 4)])

nx.draw(G, with_labels=True, font_weight='bold', connectionstyle='arc3,rad=0.1')
plt.title('Directed Graph')
plt.show()
```

```python
#Weighted Graph
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([(1, 2, {'weight': 3}), (2, 3, {'weight': 2}), (3, 1, {'weight': 1}), (1, 4, {'weight': 4})])

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_weight='bold')
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title('Weighted Graph')
plt.show()
```

```python
#Bipartite Graph
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4], bipartite=0)
G.add_nodes_from(['A', 'B', 'C'], bipartite=1)
G.add_edges_from([(1, 'A'), (1, 'B'), (2, 'B'), (3, 'C'), (4, 'A')])

pos = nx.bipartite_layout(G, [1, 2, 3, 4])
nx.draw(G, pos, with_labels=True, font_weight='bold')
plt.title('Bipartite Graph')
plt.show()
```
